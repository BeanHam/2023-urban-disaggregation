```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def statistics(low_res_name, super_res_name):
    low_res_data = np.load(f'../attributes/{low_res_name}.npy')
    low_res_geo = pd.read_csv(f'../geodata/{low_res_name}.csv')
    super_res_data = np.load(f'../attributes/{super_res_name}.npy')
    super_res_geo = pd.read_csv(f'../geodata/{super_res_name}.csv')
    linkage = np.load(f'../linkages/{low_res_name}_{super_res_name}.npy')
    
    ## calculate ratio
    ratio_matrix = np.zeros_like(super_res_data)
    for i in range(low_res_data.shape[1]):
        low_res_sub = low_res_data[:,i].reshape(-1,1)
        linkage_sub = np.where(linkage[i]==1)[0]
        super_res_sub = super_res_data[:, linkage_sub]
        ratio = super_res_sub/low_res_sub
        ratio[np.isnan(ratio)] = 0
        ratio[np.isinf(ratio)] = 0
        ratio_matrix[:, linkage_sub] = ratio
    
    ## overall variance
    plt.figure(figsize=(10,6))
    plt.plot(np.var(ratio_matrix, axis=0))
    plt.title('Hourly Variance', fontsize=18)
    plt.xlabel(f'{super_res_name}', fontsize=16)
    plt.ylabel(f'{super_res_name}/{low_res_name} Taxi Count Ratio', fontsize=16)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.savefig(f'{low_res_name}_{super_res_name}_hourly_var.png')
    pass

    ## daily variance
    plt.figure(figsize=(10,6))
    plt.plot(np.var(ratio_matrix[np.array(range(ratio_matrix.shape[0]))%24 == 0], axis=0))
    plt.title('Daily Variance', fontsize=18)
    plt.xlabel(f'{super_res_name}', fontsize=16)
    plt.ylabel(f'{super_res_name}/{low_res_name} Taxi Count Ratio', fontsize=16)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.savefig(f'{low_res_name}_{super_res_name}_daily_var.png')
    pass

    ## save geodata
    super_res_geo['hourly_var'] = np.var(ratio_matrix, axis=0)
    super_res_geo['daily_var'] = np.var(ratio_matrix[np.array(range(ratio_matrix.shape[0]))%24 == 0], axis=0)
    super_res_geo.to_csv(f'{low_res_name}_{super_res_name}_statistics.csv', index=False)
```

```python
statistics('puma', 'nta')
```

```python
statistics('nta', 'taxi')
```

```python
statistics('taxi', 'tract')
```
